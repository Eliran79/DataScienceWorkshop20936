```python
#%matplotlib notebook

import networkx as nx
```

```python
G = nx.Graph()
G.add_edge('A','B')
G.add_edge('B','C')
```

```python
nx.draw_networkx(G)
```

```python
G1 = nx.DiGraph()
G1.add_edge('B', 'A')
G1.add_edge('B', 'C')
nx.draw_networkx(G1)
```

```python
G = nx.Graph()
G.add_edge('A', 'B', weight = 6)
G.add_edge('B', 'C', weight = 60)
nx.draw_networkx(G)
```

```python
G = nx.Graph()
G.add_edge('A', 'B', relation = 'friend')
G.add_edge('B', 'C', relation = 'coworker')
G.add_edge('D', 'E', relation = 'family')
nx.draw_networkx(G)
```

```python
from networkx.algorithms import bipartite
B = nx.Graph() # no seperate class for bipartite
B.add_nodes_from(['A', 'B', 'C', 'D'], bipartite = 0)
B.add_nodes_from([1, 2, 3, 4], bipartite = 1)
B.add_edges_from([('A', 1),('A', 2), ('B', 1), ('C', 1), ('C', 2),
 ('D', 4), ('D', 3), ('D', 2)]) 
```

```python
nx.draw_networkx(B)
```

```python
# 9.5.1

B = nx.Graph()
B.add_nodes_from(['A', 'B', 'C', 'D'], bipartite = 0)
B.add_nodes_from([1, 2, 3, 4], bipartite = 1)
B.add_edges_from([('A', 1),('A', 2), ('B', 1), ('C', 1), ('C', 2), ('D', 4), ('D', 3),
('D', 2)])
nx.draw_networkx(B)
```

```python
X = set(['A', 'B', 'C', 'D'])
L = bipartite.projected_graph(B, X)
nx.draw_networkx(L)
```

```python
X = set(['A', 'B', 'C', 'D'])
L = bipartite.weighted_projected_graph(B, X)
L.edges(data=True)
nx.draw_networkx(L)
```

```python
X = set([1,2,3,4])
L = bipartite.weighted_projected_graph(B, X)
L.edges(data=True)
nx.draw_networkx(L)
```
